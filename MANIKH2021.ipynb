```python
df.drop(['Deaths / 100 Cases','Recovered / 100 Cases','Deaths / 100 Recovered','Confirmed last week','1 week change','1 week % increase','WHO Region'],axis=1,inplace=True)
```

```python
df2=df[df['New cases']>1000]
df2
```

```python
countries=df2['Country'].unique()
len(countries)
```

```python
df.head(20)
```

```python
for idx in range(0,len(countries)):
    c=df2[df2['Country']==countries[idx]].reset_index()
    plt.bar(np.arange(0,len(c)),c['Confirmed'],color='blue',label='Confirmed')
    plt.bar(np.arange(0,len(c)),c['Recovered'],color='red',label='Recovered')
    plt.bar(np.arange(0,len(c)),c['Deaths'],color='green',label='Deaths')
    plt.title(countries[idx])
    plt.xlabel('Days since first suspect')
    plt.ylabel('Number of cases')
    plt.legend()
    plt.show()
```

```python
for idx in range(0,len(countries)):
    c=df2[df2['Country']==countries[idx]].reset_index()
    plt.scatter(np.arange(0,len(c)),c['Confirmed'],color='blue',label='Confirmed')
    plt.scatter(np.arange(0,len(c)),c['Recovered'],color='red',label='Recovered')
    plt.scatter(np.arange(0,len(c)),c['Deaths'],color='green',label='Deaths')
    plt.title(countries[idx])
    plt.xlabel('Days since first suspect')
    plt.ylabel('Number of cases')
    plt.legend()
    plt.show()
```
